# JSON Files Handling 

```python
import json
dataType1 = {
        "ID": [213,214,215,216,217],
        "Name": ["Bob","Lip","Tammi","Debbie","Frank"],
        "u_ID": ["AX3","LP3","TX4","JB7","OX9"],
        "Amount": [2444, 678, 3000, 985, 2001]
}
# List of JSONS / Like Document in MongoDB .. In a Dictionary
dataType2 = {
    "ITSM Incidents":[
        {"ID":213,"UID":"AEX7", "Issue":"latency", "Priority":"P1" },
        {"ID":214,"UID":"BEJ8", "Issue":"authentication", "Priority":"P1" },
        {"ID":215,"UID":"TEX1", "Issue":"Customer level issue", "Priority":"P3" },
        {"ID":216,"UID":"EBX7", "Issue":"Infrastructure", "Priority":"P2" }
    ]
}              

# Writing the dataType2 into a JSON File
try:
    with open("SampleJSON.json", "w") as file:
        json.dump(dataType2, file, indent=4)

    print("Action Successful")
except Exception as e:
    print(f"Error writing to JSON: {e}")

```

```python
try:
    with open("SampleJSON.json", "r") as file:
        data=json.load(file)
    print(data)
    print("\n====================================\n")
    for x in data["ITSM Incidents"]:
        print(f"Record - {x["ID"]} - {x["UID"]} - {x["Issue"]} - {x["Priority"]}")

except Exception as e:
    print(f"Error while reading JSON File - {e}")
```
